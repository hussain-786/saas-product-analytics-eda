```python
sales_df = cleaned_sheets["Sales"]

print("\nColumns in Sales Sheet:")
print(sales_df.columns)
```

```python
top_products = (
    sales_df.groupby("product_name")["active_users_millions"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_products.plot(kind="bar")

plt.title("Top 10 Products by Active Users")
plt.xlabel("Product Name")
plt.ylabel("Active Users (Millions)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.histplot(
    sales_df["customer_rating"],
    bins=10,
    kde=True
)

plt.title("Customer Rating Distribution")
plt.xlabel("Customer Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(
    x="pricing_model",
    y="active_users_millions",
    data=sales_df
)

plt.title("Active Users by Pricing Model")
plt.xticks(rotation=45)
plt.show()
```
